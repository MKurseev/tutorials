## <span style="color:#0229C0">Группировка и сортировка</span>

*Первое, на чем часто я вижу ошибки:*

```python
df.groupby('season')
```

*Какая-то ерунда на первый взгляд. Но если подумать, то как по вашему должен выглядить датафрейм сгруппированный по сезонам?*


*Ответ:*

```python
from IPython.display import display
df.groupby('season').apply(display)
```

*Да, это просто разбивка на много таблиц, где каждая содержит информацию по сезону. Такое сложно запихнуть в одну таблицу, поэтому все и видят <pandas.core.groupby.generic.DataFrameGroupBy object at 0x0000022A7F64DD60>*

#### <span style="color:#990012">Поэтому группировки мы делаем, чтобы потом какое-нибудь математическое действие применить и только после этого мы сможем получить сводную таблицу с результатом.</span> 

```python
df.groupby('season')['wins'].max()
df.groupby('season')['wins'].min()
df.groupby('season')['wins'].mean()
df.groupby('season')['wins'].std()
df.groupby('season')[['wins','losses']].corr()    # эта интересная комбинация. Запустите отдельно :)
df.groupby('season')['wins'].median()
df.groupby('season')['wins'].count()
df.groupby('season')['wins'].sum()     # смотрим, сколько всего было побед в каждом сезоне

## группируем по сезонам, а потом смотрим какое-то математическое действие для конкрутного столбца
```

#### <span style="color:#990012">Можно группировать по нескольким столбцам. (тут не забываем, что перечисление столбцов идет в list)</span> 

```python
df.groupby(['season','team'])['points'].sum().head(50)
```

*Тут мы смотрим, сколько в каждом сезоне набрала каждая команда очков.*

 ## <span style="color:#0229C0">sort_values()</span>

#### <span style="color:#990012">Еще нам нужно знать сортировку. Тут все просто. Когда получили нужный нам ответ, дописываем **sort_values()** </span>

```python
df.groupby('season')['wins'].sum().sort_values(ascending=False) # ascending = от макс. к мин или наоборот
```

*sort_values можно использовать без предвариельной обработки датафрейма. Тогда еще нужно указывать, по какому столбцу сортировать*

```python
df[['team','season','games','points']].sort_values(by=['season','points'], ascending=True)
# отсортировали по сезонам, а потом еще и по очкам от меньшего к большему 
```

*Если вы хотите, например, сортировать по годам от первого к последнему, но при этом количество очков от максимума к мин, то просто укажите это в ascending*

```python
df[['team','season','games','points']].sort_values(by=['season','points'],ascending=[True,False])
```

## <span style="color:#0229C0">Интервалы</span>
#### <span style="color:#990012">Для них есть функция cut(). В качестве аргументов в нее подаем данные и интервал. </span>

```python
print(df['points'])    # возьмем столбец кол-ва очков
```

```python
pd.cut(df['points'],5)   # Разбили диапазон всевозможных очков на 5 равных интервалов. Сами интервалы указаны внизу.
                         # по ней мы видем в каком диапазоне лежит кол-во очков в каждой ячейке столбца
```

#### Categories (5, interval[float64]): [(35.908, 54.4] < (54.4, 72.8] < (72.8, 91.2] < (91.2, 109.6] < (109.6, 128.0]]

*А теперь скомбинируем получившийся столбец с функцией value_counts. Получим уникальные значения интервалов и сколько раз они встречались.*

```python
pd.cut(df['points'],5).value_counts(sort=False)
```

*Можно не просто указывать кол-во интервалов, а явно их указывать, что может быть удобнее*

```python
pd.cut(df['points'], [35,50,75,100,130]).value_counts(sort=False)
```

## <span style="color:#0229C0">Задача</span>
#### <span style="color:#990012">Для каждого сезона найдите команду, которая пропустила наименьшее число голов.</span>

*Какая идея решения?*


*Найдем индексы строчек, которые нам нужны. А потом Из всей таблицы оставем только те строчки, чьи индексы совпадают с найденными.*

```python
imax = df.groupby(['season'])['goals_against'].apply(lambda x: x.idxmin())
imax   # для каждого сезона мы нашли индекс строчки, у которой goals_against наименьшее
```

```python
imax.values     # индексы, найденные ранее
```

```python
df.index     # все индексы df
```

```python
df2 = df[df.index.isin(imax.values)]   # наложили условие, что индексы df только те которые нашли ранее.
                                       # функция isin() - оставить то , что лежит внутри меня.    
df2[['season','team','games','wins','losses','goals_against']]            
```

```python
df2[['season','team','goals_against']].sort_values(by='season')
                        # оставим только нужные столбцы и отсортируем по сезонам.
```
